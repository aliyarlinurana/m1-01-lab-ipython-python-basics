```python
values = [10, 20, 30]
alias = values
alias.append(40)
```

```python
print("values: ",values)
print("alias: ",alias)
```

```python
values_copy=values.copy()
values_copy.append(50)
print("values: ",values)
print("copy of values: ",values_copy)
```

```python
record={"name":"Nurana","Surname":"Aliyarli","age":20}
record_alias=record
record_alias["city"]="Baku"
```

```python
print("record: ",record)
print("record_alias: ",record_alias)
```

```python
record_copy=record.copy()
record_copy["university"]="Khazar"
print("record: ",record)
print("record_copy: ",record_copy)
```

When I use an alias, both names point to the same object in memory,so if I change one, the other changes too. But when I use .copy(), I create a new, independent object, so the original stays safe.
This is also same for functions,if I pass a list into a function, the function can accidentally change my original data because it uses the same reference.To prevent this, I should pass a copy of the list to the function instead.

```python
user_events= [
    {"user_id":1, "event_type":"login", "duration_seconds":60},
    {"user_id":2, "event_type":"view", "duration_seconds":120},
    {"user_id":3, "event_type":"click", "duration_seconds":-30},  
    {"user_id":4, "event_type":"view", "duration_seconds":75},
    {"user_id":5, "event_type":"logout", "duration_seconds":45},
    {"user_id":6, "event_type":"login", "duration_seconds":"none"},
    {"user_id":7, "event_type":"click", "duration_seconds":180},
    {"user_id":8, "event_type":"view", "duration_seconds":240},
    {"user_id":9, "event_type":"login", "duration_seconds":-90},
    {"user_id":10, "event_type":"view", "duration_seconds":600},
    {"user_id":11, "event_type":"click", "duration_seconds":15},
    {"user_id":12, "event_type":"logout", "duration_seconds":50}]

cleaned_events = []

for event in user_events:
    duration = event["duration_seconds"]
    
    if type(duration) in (int, float) and duration >= 0:
        event["duration_minutes"] = duration / 60
        cleaned_events.append(event)
        
print("cleaned count:", len(cleaned_events))

for event in cleaned_events:
    if not all(key in event for key in ["user_id", "event_type", "duration_seconds", "duration_minutes"]):
        print("missing key!")

print(cleaned_events[1])
print(cleaned_events[2])

```

we build a dictionary of event counts by event_type

```python
event_counts = {}

for event in cleaned_events:
    event_type = event["event_type"]
    
    if event_type in event_counts:
        event_counts[event_type] += 1
    else:
        event_counts[event_type] = 1

print("Event counts:", event_counts)
```

then we build a dictionary of average duration_minutes by event_type. 

```python
duration_totals = {}
event_numbers = {}

for event in cleaned_events:
    event_type = event["event_type"]
    duration = event["duration_minutes"]
    
    if event_type in duration_totals:
        duration_totals[event_type] += duration
        event_numbers[event_type] += 1
    else:
        duration_totals[event_type] = duration
        event_numbers[event_type] = 1

avg_duration = {}
for event_type in duration_totals:
    avg_duration[event_type] = duration_totals[event_type] / event_numbers[event_type]

print("average duration:", avg_duration)
```

we compute number of unique users

```python
unique_users = set()
for event in cleaned_events:
    unique_users.add(event["user_id"])

print("number of unique users:", len(unique_users))
```

we check the the one validation here

```python
total_count = sum(event_counts.values()) 
if total_count == len(cleaned_events):
    print("validation passed")
else:
    print("validation failed")
```

```python
import io
import csv

csv_rows = [["metric", "key", "value"]]

#one row per event type for the counts
for event_type, count in event_counts.items():
    csv_rows.append(["count", event_type, count])

#one row per event type for the averages
for event_type, avg in avg_duration.items():
    csv_rows.append(["average", event_type, avg])

#final row for the unique user count
csv_rows.append(["unique_users", "all", len(unique_users)])

output = io.StringIO()
writer = csv.writer(output)
writer.writerows(csv_rows)
csv_string = output.getvalue()

print(csv_string)
```
